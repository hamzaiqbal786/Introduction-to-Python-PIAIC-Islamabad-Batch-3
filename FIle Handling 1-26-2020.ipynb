- With Block automatically close the connectivity with file and release the resources

```python
with open("hello.txt") as file:
    content = file.read()
    print(content)
```

```python
with open("hello.txt") as file:
    content = file.readline()
    print(content)
```

# Write() 

```python
with open("FileTesting.txt",mode='w') as file:
    content = file.write("Hello World")
    print(content)
```

# Append()

```python
with open("FileTesting.txt",mode='a') as file:
    content = file.write("Appending")
    print(content)
```

```python
with open("hello123.txt",mode='a+') as file:
    file.write("Appending")
    file.seek(0)
    content = file.readlines()
    print(content)
```
